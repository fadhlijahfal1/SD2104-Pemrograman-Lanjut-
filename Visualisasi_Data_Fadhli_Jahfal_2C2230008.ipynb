1.Histogram Distribusi Nilai Matematika

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
plt.hist(df['nilai_mat'], bins=10, color='red', edgecolor='yellow')
plt.title('Distribusi Nilai Matematika', fontsize=14)
plt.xlabel('Nilai Matematika', fontsize=10)
plt.ylabel('Frekuensi', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Keterangan :
Barplot ini menunjukkan bahwa sebagian besar mahasiswa memiliki nilai matematika di rentang 60-80.

2.Diagram Batang Horizontal Banyaknya Mahasiswa masing-masing Fakultas.

```python
fakultas_count = df['fakultas'].value_counts()


plt.figure(figsize=(8, 6))
plt.barh(fakultas_count.index, fakultas_count.values, color='magenta', edgecolor='orange')
plt.title('Jumlah Mahasiswa per Fakultas', fontsize=14)
plt.xlabel('Jumlah Mahasiswa', fontsize=12)
plt.ylabel('Fakultas', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

3.Diagram Pie Banyaknya mahasiswa masing-masing fakultas

```python
plt.figure(figsize=(8, 8))
plt.pie(fakultas_count, labels=fakultas_count.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title('Proporsi Mahasiswa per Fakultas', fontsize=12)
plt.show()
```
